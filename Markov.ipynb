```python
df_path= df.groupby('cookie')['channel'].agg(lambda x: x.tolist()).reset_index()
df_path= df_path.rename(columns={'channel':'path'})
df_path.head()
```

```python
df_last_interaction= df.drop_duplicates('cookie',keep='last')[['cookie',
                                                               'conversion',
                                                               'conversion_value']]

df_last_interaction.head()
```

```python
df_cust_journey =pd.merge(df_path, df_last_interaction, how='left', on='cookie')
df_cust_journey.head()
```

```python
print('total conversion = ', sum(df_cust_journey['conversion']))
print('conversion rate = ',sum(df_cust_journey['conversion'])/len(df)*100 )
```

```python
def path_creation(df):
  str1=''
  for i in df['path']:
    str1 += i + ' > '
  return str1[:-3]
```

```python
df_cust_journey['path']= df_cust_journey.apply(path_creation, axis=1)
df_cust_journey.head()
```
